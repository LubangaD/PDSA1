**We create a rating map to for the attributes or columns**

```python
rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

columns_to_map = [
    'Provide a self evaluation of your work in the group. [Contributions & Attitude]',
    'Provide a self evaluation of your work in the group. [Cooperation with Others]',
    'Provide a self evaluation of your work in the group. [Focus, Commitments]',
    'Provide a self evaluation of your work in the group. [Team Role Fulfillment]',
    'Provide a self evaluation of your work in the group. [Ability to Communicate]',
    'Provide a self evaluation of your work in the group. [Accuracy]',
    'Group Activity Rating [Contributions & Attitude]',
    'Group Activity Rating [Cooperation with Others]',
    'Group Activity Rating [Focus, Commitments]',
    'Group Activity Rating [Team Role Fulfillment]',
    'Group Activity Rating [Ability to Communicate]',
    'Group Activity Rating [Accuracy]',
    'Group Activity Rating [Contributions & Attitude].1',
    'Group Activity Rating [Cooperation with Others].1',
    'Group Activity Rating [Focus, Commitments].1',
    'Group Activity Rating [Team Role Fulfillment].1',
    'Group Activity Rating [Ability to Communicate].1',
    'Group Activity Rating [Accuracy].1',
    'Group Activity Rating [Contributions & Attitude].2',
    'Group Activity Rating [Cooperation with Others].2',
    'Group Activity Rating [Focus, Commitments].2',
    'Group Activity Rating [Team Role Fulfillment].2',
    'Group Activity Rating [Ability to Communicate].2',
    'Group Activity Rating [Accuracy].2'
]

# Apply mapping to specified columns
for column in columns_to_map:
    data[column] = data[column].map(rating_map)
```

```python
# columns with missing values
data.isnull().sum()
```

```python
#Handle missing values by filling with 1
data = data.fillna(value=1)
```

```python
data.isnull().sum()
```

**Add column rating by finding the mean**

```python
mean_columns = data.columns[2:8]
data['Rating'] = data[mean_columns].mean(axis=1)
# Convert the 'Rating' column to float with two decimal places
data['Rating'] = data['Rating'].round(2).astype(float) # part of data cleaning making sure the rating column has a floating data type
```

- The relationship between the predictor and response variables is linear- For example if the our target variable is to predict number of customers that use the system, the we would assume that it is linear to variables such as time, diseases.

- The observations or the variables in the data must be independent of each other. This means that the value of one observation should not be influenced by the value of another observation.

- Independent variables  should not be highly correlated with each other as Multicollinearity can lead to unstable estimates of the regression coefficients.

Outlier  is a data point that significantly differs from other data points in a dataset.
An outlier can be found above or below the range of other data points  and can be clearly noted.

d)

Outliers can be identified using the following approaches;

- Using graphical methods- graphical representation such as Box plot and scatter plots are used to identify outliers. Outliers can be identified as individual points outside the whiskers of the box plot. For the scatter plot, the data points far away from the scatter are treated as outliers.


- Z-score method is a statistical method that calculates z-score of each data point, which represents how many standard deviations away a data point is from the mean. Z-score method takes data points greater 2 or 3 as outliers.

- IQR method: IQR (interquartile range) method finds the range between  the first quartile (Q1) and the third quartile (Q3) of the dataset and identifies Outliers  as data points that are more than 1.5 times the IQR away from Q1 or Q3.
